## Define constraints

The purpose of these masks is to isolate regions in the PSF that we either want to darken or brighten. We use these masks in the objective function.

**Dark annulus**
Reduce light in the ring.

**Central peak**
Maintain light at the very center of the PSF to maintain a high contrast ratio.

```python
# Dark annulus
r = radius_meshgrid(ISZ)
minr = 24.6; maxr = 57.4
mask = (r<(maxr*2/(ISZ*2)))*(r>(minr*2/(ISZ*2)))

# Central peak
light_mask = (r < (2 * 2/(ISZ * 2)))

plt.figure(figsize = (10, 4))
plt.subplot(1, 2, 1); plt.imshow(mask, interpolation = 'none');
plt.colorbar(); plt.title('Dark annulus')
plt.subplot(1, 2, 2); plt.imshow(light_mask, interpolation = 'none'); plt.colorbar();
plt.title('Bright central peak')
plt.show()
```

## Define objective function

```python
def objective_function(coeffs, basis, support, mask, light_mask, PSZ, c1, c2):
    wf = np.dot(basis, coeffs).reshape(PSZ, PSZ)

    psf, _ = get_psf(wf, support, psz = PSZ)
    
    dark_peak = np.max(mask*psf)
    max_peak = np.max(light_mask * psf)
    
    return c1 * dark_peak + c2 * max_peak
```

## Run gradient descent

```python
results = []
```

```python
def gradient_descent(coeffs, epochs, learning_rate, results, c1, c2, save_rate = 100):
    grad_obj = (grad(objective_function))
    for i in range(epochs):
        coeffs -= lr*grad_obj(coeffs, basis, support, mask, light_mask, PSZ, c1, c2)
        
        if i % save_rate == 0:
            results.append(coeffs)
            
    results.append(coeffs)
```

```python
lr = 1e-4; epochs = 50; gradient_descent(coeffs, epochs, lr, results, c1 = 1, c2 = -1e2)
lr = 1e-4; epochs = 200; gradient_descent(results[-1], epochs, lr, results, c1 = 2e3, c2 = -1e2)
lr = 1e-5; epochs = 250; gradient_descent(results[-1], epochs, lr, results, c1 = 5e3, c2 = -1e2)
lr = 1e-6; epochs = 250; gradient_descent(results[-1], epochs, lr, results, c1 = 5e3, c2 = -1e2)

lr = 1e-5; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 1e4, c2 = -1e2)
lr = 1e-5; epochs = 1500; gradient_descent(results[-1], epochs, lr, results, c1 = 2e4, c2 = -1e2)
lr = 1e-5; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 3e4, c2 = -1e2)

lr = 1e-5; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 4e4, c2 = -1e2)
lr = 1e-6; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 4e4, c2 = -1e2)
lr = 1e-5; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 6e4, c2 = -1e2)
lr = 1e-6; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 6e4, c2 = -1e2)
lr = 1e-5; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 8e4, c2 = -1e2)
lr = 1e-6; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 8e4, c2 = -1e2)

lr = 1e-6; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 1e5, c2 = -1e2)
lr = 1e-7; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 1e5, c2 = -1e2)
lr = 1e-6; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 2e5, c2 = -1e2)
lr = 1e-7; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 2e5, c2 = -1e2)
```

# **Pixel basis**

## Initialize

```python
results = np.load('corona_mode_coeffs_paper.npy')
coeffs = results[-1]
```

```python
wf = np.dot(basis, coeffs).reshape(PSZ, PSZ)

qwfA = wf[:Q, :Q] # quater wavefront

qwfB, qwfC, qwfD = np.rot90(qwfA, k = 3), np.rot90(qwfA, k = 2), np.rot90(qwfA, k = 1)
top, bottom = np.hstack((qwfA, qwfB)), np.hstack((qwfD, qwfC))
wf = np.vstack((top, bottom))

coeffs = qwfA.reshape(-1)
```

## Define new objective function

Since we have 4-fold symmetry, we only need to optimize pixels in one quadrant and then tile these pixels to fill the pupil.

```python
def objective_function(coeffs, basis, support, mask, light_mask, Q, c1, c2):
    
    qwfA = coeffs.reshape(Q, Q)
    
    qwfB, qwfC, qwfD = np.rot90(qwfA, k = 3), np.rot90(qwfA, k = 2), np.rot90(qwfA, k = 1)
    top, bottom = np.hstack((qwfA, qwfB)), np.hstack((qwfD, qwfC))
    wf = np.vstack((top, bottom))

    psf, _ = get_psf(wf, support, psz = PSZ)
    
    dark_peak = np.max(mask*psf)
    max_peak = np.max(light_mask * psf)
    
    return c1 * dark_peak + c2 * max_peak
```

```python
results = []
```

```python
def gradient_descent(coeffs, epochs, learning_rate, results, c1, c2, save_rate = 100):
    grad_obj = (grad(objective_function))
    for i in range(epochs):
        coeffs -= lr*grad_obj(coeffs, basis, support, mask, light_mask, Q, c1, c2)

        qwfA = coeffs.reshape(Q, Q)
        qwfB, qwfC, qwfD = np.rot90(qwfA, k = 3), np.rot90(qwfA, k = 2), np.rot90(qwfA, k = 1)
        top, bottom = np.hstack((qwfA, qwfB)), np.hstack((qwfD, qwfC))
        
        wf = np.vstack((top, bottom))
        psf, _ = get_psf(wf, support, psz = PSZ)
        
        if i % save_rate == 0:
            results.append(coeffs)
            
    results.append(coeffs)
```

```python
lr = 1; epochs = 500; gradient_descent(coeffs, epochs, lr, results, c1 = 2e5, c2 = -1e2)
lr = 0.1; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 2e5, c2 = -1e2)

lr = 1; epochs = 2500; gradient_descent(results[-1], epochs, lr, results, c1 = 5e5, c2 = -1e2)
lr = 0.1; epochs = 2000; gradient_descent(results[-1], epochs, lr, results, c1 = 5e5, c2 = -1e2)
lr = 0.01; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 5e5, c2 = -1e2)
lr = 1; epochs = 500; gradient_descent(results[-1], epochs, lr, results, c1 = 1e6, c2 = -1e2)
lr = 0.1; epochs = 3500; gradient_descent(results[-1], epochs, lr, results, c1 = 1e6, c2 = -1e2)

lr = 0.1; epochs = 6000; gradient_descent(results[-1], epochs, lr, results, c1 = 5e6, c2 = -1e2)
lr = 0.1; epochs = 1000; gradient_descent(results[-1], epochs, lr, results, c1 = 1e7, c2 = -1e2)
lr = 0.01; epochs = 2000; gradient_descent(results[-1], epochs, lr, results, c1 = 1e7, c2 = -1e2)
lr = 0.001; epochs = 1000; gradient_descent(results[-1], epochs, lr, results, c1 = 1e7, c2 = -1e2)
lr = 0.01; epochs = 1000; gradient_descent(results[-1], epochs, lr, results, c1 = 5e7, c2 = -1e2)
lr = 0.001; epochs = 2000; gradient_descent(results[-1], epochs, lr, results, c1 = 5e7, c2 = -1e2)

lr = 0.01; epochs = 1000; gradient_descent(results[-1], epochs, lr, results, c1 = 1e6, c2 = -1e2)
```
